```python
import pandas as pd
import numpy as np
```

```python
book=pd.read_csv("book.csv")
book
```

```python
book.rename(columns={'User.ID':'userId','Book.Title':'BookTitle','Book.Rating':'BookRating'},inplace=True)
```

```python
book
```

```python
book.nunique()
```

```python
book.isnull().sum()
```

```python
book_df=book.pivot_table(index="userId",
                     columns="BookTitle",
                     values="BookRating",aggfunc="mean")
```

```python
book_df
```

```python
book_df.fillna(0,inplace=True)
```

```python
book_df
```

```python
from scipy.spatial.distance import cosine
from sklearn.metrics import pairwise_distances
```

```python
similarity=1-pairwise_distances(book_df.values,metric='cosine')
```

```python
similarity
```

```python
similarity_df=pd.DataFrame(similarity)
```

```python
similarity_df
```

```python
similarity_df.index=list(book_df.index)
similarity_df.columns=list(book_df.index)
```

```python
similarity_df
```

```python
np.fill_diagonal(similarity,0)
```

```python
similarity_df
```

```python
similarity_df.sort_values([8],ascending=False).head(10).index
```

```python


def get_recom(cus_ID):
 
        tem=list(similarity_df.sort_values([cus_ID],ascending=False).head(100).index)
        book_list=[]
        for i in tem:
            book_list=book_list+list(book[book['userId']==i]["BookTitle"])
        return set(book_list)-set(book[book['userId']==cus_ID]['BookTitle'])
      
        
```

```python
get_recom(8)
```
